MODEL IMPORTING

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']]
y = df['Species']
```

```python
X_train, X_test, y_train,y_test = train_test_split(X,y,test_size =0.2)
```

CLASSIFICATION ALGORITHM APPLYING

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```python
# Create an SVM classifier
SVC_classification = SVC(kernel='linear', C=1)
SVC_classification.fit(X_train, y_train)
y_pred = SVC_classification.predict(X_test)
y_pred
```

```python
# Calculate and print the accuracy of the model
SVC_acc = accuracy_score(y_test, y_pred)
print("Accuracy:", SVC_acc)
```

```python
# Print the classification report, which includes precision, recall, and F1-score for each class
print(classification_report(y_test, y_pred))
```

LOGISTIC REGRESSION CLASSIFICATION

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR_Classification = LogisticRegression(max_iter=1000)
LR_Classification.fit(X_train, y_train)
y_pred1 = LR_Classification.predict(X_test)
y_pred1
```

```python
#During the classification we check the accuracy of our model
LR_acc = accuracy_score(y_test, y_pred1)
print("Accuracy:", LR_acc)
```

```python
print(classification_report(y_test, y_pred1))
```
